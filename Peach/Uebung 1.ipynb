```python
#Spreizung
#Die Spreizung ist eine affine Punktoperation
#Nachteil:
#  In der Praxis nicht robust
#  Nur unter optimalen Bedingungen anwendbar

import numpy as np
from numpy import *

# Algorithmus 11
# Die Spreizung berechnet in einem Bild zunächst die
# minimale Intensität min und die maximale Intensität max
def BerechneMinMax(I):
    min=q
    max=0
    u=0
    v=0    
    while v<4:
        while u<3:
            if I[u,v]<min:
                min=I[u,v]
            elif I[u,v]>max:
                max=I[u,v]
            u=u+1        
        v=v+1
        u=0
    return min,max

# Algorithmus 8
def AffinePunktoperatioin(I,a,b):
    v=0
    u=0
    while v<4:
        while u<3:
            g=a*I[u,v]+b                   #Wegen Python3+ round() nicht wie gewuenscht funktioniert
            g_=round(a*I[u,v]+b)
            if (g-g_)>=0.5:
                g=g_+1
            else:
                g=g_
            if g<0:
                I[u,v]=0
            elif g>q:
                I[u,v]=q
            else:
                I[u,v]=g
            u=u+1
        v=v+1
        u=0
    return I

# Algorithmus 12
# Anschließend wird das Intervall [min, max] linear auf
# [0, 255] abgebildet und es werden entsprechend die
# Intensitäten im Bild angepasst
def Spreize(I):
    minmax=BerechneMinMax(I)
    Min=minmax[0]
    Max=minmax[1]
    I_=I
    if Min==Max:
        print('Bild I ist homogen')
    else:
        a=q/(Max-Min)
        b=-q*Min/(Max-Min)
        I_=AffinePunktoperatioin(I_,a,b)
    return I_



q=255
B=mat([[180,80,140,110],
       [0,110,140,80],
       [255,110,180,140]])
B_=Spreize(B)
print('B_=\n',B_)
```

```python
# Histogrammdehnung
# Die Histogrammdehnung ist eine Verbesserung der Spreizung
# Anstatt der minimalen bzw. maximalen Intensität werden
# Quantile verwendet, um tatsächliche Maxima im Histogramm zu erkennen
# Die Histogrammdehnung ist eine affine Punktoperation

import numpy as np
from numpy import *

# Algorithmus 8
def AffinePunktoperatioin(I,a,b):
    v=0
    u=0
    while u<3:
        while v<4:
            g=a*I[u,v]+b                   #Wegen Python3+ round() nicht wie gewuenscht funktioniert
            g_=round(a*I[u,v]+b)
            if (g-g_)>=0.5:
                g=g_+1
            else:
                g=g_
#             print(g_)    
            if g<0:
                I[u,v]=0
            elif g>q:
                I[u,v]=q
            else:
                I[u,v]=g
#             print(I[u,v])
            v=v+1
        u=u+1
        v=0
    return I

# Algorithmus 14
def BrechneQuantil(Ha,p):
    x=0
    while x<(q+1):
        if Ha[x]>=p*Ha[q]:
            v=x
            return v
        else:
            x=x+1

# Algorithmus 13
def BrechneAkkumuliertesHistogramm(H):
    Ha=np.empty((q+1))
    Ha[0]=H[0]
    x=1
    while x<(q+1):
        Ha[x]=Ha[x-1]+H[x]
        x=x+1
    return Ha

# Algorithmus 9
def BrechneGrauwertHistogramm(I):
    x=0
    u=0
    v=0
    H=np.empty((q+1))
    while x<(q+1):
        H[x]=0
        x=x+1
    while v<4:
        while u<3:
            H[I[u,v]]=H[I[u,v]]+1
            u=u+1
        v=v+1
        u=0
    return H

# Algorithmus 15
def HistogrammDehnung(I,pmin,pmax):
    H=BrechneGrauwertHistogramm(I)
    Ha=BrechneAkkumuliertesHistogramm(H)
    min=BrechneQuantil(Ha,pmin)
    max=BrechneQuantil(Ha,pmax)
    if min==max:
        print('Bild I ist homogen')
        return None
    else:
        a=q/(max-min)
        b=-q*min/(max-min)
#         print(a,b)
        I_=AffinePunktoperatioin(I,a,b)
        return I_

q=255
B=mat([[180,80,140,110],
       [0,110,140,80],
       [255,110,180,140]])
B_=HistogrammDehnung(B,0.1,0.9)
print('B_=\n',B_)

# Onlinefrage Nr. 3:
# Wie groß ist die Summe aller Eintr¨age der neuen Bildmatrix aus Teilaufgabe 3.b?
u=0
v=0
s=0
while u<3:
    while v<4:
        s=s+B[u,v]
        v=v+1
    u=u+1
    v=0
print(s)
```
